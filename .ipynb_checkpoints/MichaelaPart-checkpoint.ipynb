# SVM on Iris Dataset

Below are various forms of SVM on the Iris dataset. The first is linear SVM, followed by polynomial SVM, then Gaussian SVM.

```python
import numpy as np
from sklearn import datasets 
from sklearn import svm
from sklearn.svm import SVC 
from sklearn.model_selection import train_test_split   
from matplotlib import pyplot as plt


iris_dataset = datasets.load_iris()           # Import iris data 

X = iris_dataset['data']                      # X is the data
Y = iris_dataset['target']                    # Y is the labels
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.30)   # 30% of data to test on
```

### PCA of Iris Dataset

```python
# Computation of the mean of training data
X_train_T = X_train.T
mean_vec = X_train_T.mean(axis = 1)
mean_mat = np.tile(mean_vec.reshape(X_train_T.shape[0],1),[1,X_train_T.shape[1]])

# Centered training data matrix
train_centered_X = X_train_T - mean_mat

# Normalize training data matrix
standard_deviation = np.array([[np.std(i)] for i in train_centered_X])
train_normal_X = train_centered_X/standard_deviation

# Compute two-dimensional PCA features of the centered and normalized dataset
U, s, Vh = np.linalg.svd(train_normal_X)    # Singular value decomposition 

# Matrix of the top two principal components of the centered, normalized training data matrix
U_2 = U[:,0:2]

# PCA-based features of the centered, normalized data using the top-two principal components
X_tilde_train = U_2.T@train_normal_X

# Computation of the mean of test data
X_test_T = X_test.T
mean_mat = np.tile(mean_vec.reshape(X_test_T.shape[0],1),[1,X_test_T.shape[1]])

# Centered test data matrix
test_centered_X = X_test_T - mean_mat

# Normalize test data matrix
standard_deviation = np.array([[np.std(i)] for i in test_centered_X])
test_normal_X = test_centered_X/standard_deviation

# PCA-based features of the centered, normalized data using the top-two principal components
X_tilde_test = U_2.T@test_normal_X
```

### Plotting Iris Data

```python
# Scatter plot of 2D PCA features
plt.figure(figsize=(10,10))
setosa = plt.scatter(X_tilde_train.T[np.where(Y_train==0),0], X_tilde_train.T[np.where(Y_train==0),1],c='r',marker='o')
versicolor = plt.scatter(X_tilde_train.T[np.where(Y_train==1),0],X_tilde_train.T[np.where(Y_train==1),1],c='g',marker='o')
virginica = plt.scatter(X_tilde_train.T[np.where(Y_train==2),0],X_tilde_train.T[np.where(Y_train==2),1],c='b',marker='o')
plt.xlabel("Principal Component 1", fontsize = 15)
plt.ylabel("Principal Component 2", fontsize = 15)
plt.title("Two-dimensional PCA of Iris Dataset", fontsize = 15)
plt.legend((setosa, versicolor, virginica), ("Setosa", "Versicolor", "Virginica"))
plt.show()
```

### K-Fold of Iris Data

```python
k = round(105/5)                      # Round to find length/5 for K=5
f = len(X_tilde_train.T)              # Length of samples

traindata_1 = X_tilde_train[:,k:f]    # First partition
trainlabels_1 = Y_train[k:f]
testdata_1 = X_tilde_train[:,0:k]
testlabels_1 = Y_train[0:k]

traindata_2 = np.column_stack((X_tilde_train[:,0:k],X_tilde_train[:,(2*k):]))      # Second partition
trainlabels_2 = np.append(Y_train[:k],Y_train[(2*k):])
testdata_2 = X_tilde_train[:,k:(2*k)]
testlabels_2 = Y_train[k:(2*k)]

traindata_3 = np.column_stack((X_tilde_train[:,0:(2*k)],X_tilde_train[:,(3*k):]))  # Third partition
trainlabels_3 = np.append(Y_train[:(2*k)],Y_train[(3*k):])
testdata_3 = X_tilde_train[:,(2*k):(3*k)]
testlabels_3 = Y_train[(2*k):(3*k)]

trainlabels_4 = np.append(Y_train[:(3*k)],Y_train[(4*k):])                         # Fourth partition
traindata_4 = np.column_stack((X_tilde_train[:,0:(3*k)],X_tilde_train[:,(4*k):]))
testdata_4 = X_tilde_train[:,(3*k):(4*k)]
testlabels_4 = Y_train[(3*k):(4*k)]

trainlabels_5 = Y_train[:(4*k)]                                                    # Fifth partition
traindata_5 = X_tilde_train[:,0:(4*k)]
testdata_5 = X_tilde_train[:,(4*k):]
testlabels_5 = Y_train[(4*k):]

def calculateError(pred,labels):             # Error calculation function
    count = 0;
    for i in range(len(pred)):
        if (pred[i] != labels[i]):
            count = count +1
    percentage = count/(len(pred))
    return percentage
```

### Linear SVM of Iris Dataset

```python
def LinearSVM(X_tilde_train, Y_train, X_tilde_test, Y_test, C_val):
    """
    Linear SVM classifier that returns either label "0", "1", or "2"
    Inputs:
        X_train: Training data matrix
        Y_train: Labels of training data matrix
        X_test: Test data matrix
        Y_test: Labels of test data matrix
    Outputs: 
        Y_pred: Predicted label of test data matrix
    """
    svclassifier = SVC(C = C_val, kernel = 'linear')        # Linear SVM
    svclassifier.fit(X_tilde_train.T, Y_train)              # Fit the data
    Y_pred_Linear = svclassifier.predict(X_tilde_test.T)    # Make predictions on test data
    return Y_pred_Linear

def bestC(traindata_1,trainlabels_1,testdata_1, testlabels_1, c1, labels1,
          traindata_2,trainlabels_2,testdata_2, testlabels_2, c2, labels2,
          traindata_3,trainlabels_3,testdata_3, testlabels_3, c3, labels3,
          traindata_4,trainlabels_4,testdata_4, testlabels_4, c4, labels4,
          traindata_5,trainlabels_5,testdata_5, testlabels_5, c5, labels5):
    prediction_1 = LinearSVM(traindata_1,trainlabels_1,testdata_1, testlabels_1, c1) # First partition predictions
    prediction_2 = LinearSVM(traindata_2,trainlabels_2,testdata_2, testlabels_2, c2) # Second partition predictions
    prediction_3 = LinearSVM(traindata_3,trainlabels_3,testdata_3, testlabels_3, c3) # Third partition predictions
    prediction_4 = LinearSVM(traindata_4,trainlabels_4,testdata_4, testlabels_4, c4) # Fourth partition predictions
    prediction_5 = LinearSVM(traindata_5,trainlabels_5,testdata_5, testlabels_5, c5) # Fifth partition predictions

    errors = list()
    errors.append(calculateError(prediction_1,testlabels_1))     # Errors for first partition
    errors.append(calculateError(prediction_2,testlabels_2))     # Errors for second partition
    errors.append(calculateError(prediction_3,testlabels_3))     # Errors for third partition
    errors.append(calculateError(prediction_4,testlabels_4))     # Errors for fourth partition
    errors.append(calculateError(prediction_5,testlabels_5))     # Errors for fifth partition

    min_error = np.min(errors)
    min_pred = errors.index(min(errors))
    if(min_pred==0):
        best_C = c1
    if(min_pred==1):
        best_C = c2
    if(min_pred==2):
        best_C = c3
    if(min_pred==3):
        best_C = c4
    if(min_pred==4):
        best_C = c5
    return min_error, best_C
      
    
c1 = 1
c2 = 5
c3 = 10
c4 = 20
c5 = 100

min_error, best_C = bestC(traindata_1,trainlabels_1,testdata_1, testlabels_1, c1, testlabels_1,
          traindata_2,trainlabels_2,testdata_2, testlabels_2, c2, testlabels_2,
          traindata_3,trainlabels_3,testdata_3, testlabels_3, c3, testlabels_3,
          traindata_4,trainlabels_4,testdata_4, testlabels_4, c4, testlabels_4,
          traindata_5,trainlabels_5,testdata_5, testlabels_5, c5, testlabels_5)
 
print("Best C from K-Fold = ", best_C)

# Save Linear classifier predictions
Y_pred_Linear = np.array([LinearSVM(X_tilde_train, Y_train, X_tilde_test, Y_test, best_C)])

# Accuracy of the predicted labels
acc_label1 = np.equal(Y_pred_Linear,Y_test)

# Plot training data
plt.figure(figsize=(10,10))
setosa = plt.scatter(X_tilde_train.T[np.where(Y_train==0),0], X_tilde_train.T[np.where(Y_train==0),1],c='r',marker='o')
versicolor = plt.scatter(X_tilde_train.T[np.where(Y_train==1),0],X_tilde_train.T[np.where(Y_train==1),1],c='g',marker='o')
virginica = plt.scatter(X_tilde_train.T[np.where(Y_train==2),0],X_tilde_train.T[np.where(Y_train==2),1],c='b',marker='o')
plt.xlabel("Principal Component 1", fontsize = 15)
plt.ylabel("Principal Component 2", fontsize = 15)
plt.title("Two-dimensional PCA", fontsize = 15)

# Plot for test data
correct = plt.scatter(X_tilde_test.T[np.where(acc_label1==True),0],X_tilde_test.T[np.where(acc_label1==True),1],c='y',marker='s')
incorrect = plt.scatter(X_tilde_test.T[np.where(acc_label1==False),0],X_tilde_test.T[np.where(acc_label1==False),1],c='k',marker='^')
plt.legend((setosa, versicolor, virginica, correct, incorrect), 
               ("Setosa", "Versicolor", "Virginica", "Correctly Classified", "Incorrectly Classified"))
plt.show()
```

### Polynomial SVM on Iris Dataset

```python
def PolySVM(X_tilde_train, Y_train, X_tilde_test, Y_test, Deg_val):
    """
    Poly SVM classifier that returns either label "0", "1", or "2"
    Inputs:
        X_train: Training data matrix
        Y_train: Labels of training data matrix
        X_test: Test data matrix
        Y_test: Labels of test data matrix
    Outputs: 
        Y_pred: Predicted label of test data matrix
    """
    svclassifier = SVC(kernel = 'poly', degree = Deg_val) # Polynomial SVM
    svclassifier.fit(X_tilde_train.T, Y_train)            # Fit the data
    Y_pred_Poly = svclassifier.predict(X_tilde_test.T)    # Make predictions on test data
    return Y_pred_Poly



def bestDeg(traindata_1,trainlabels_1,testdata_1, testlabels_1, Deg_val1, labels1,
          traindata_2,trainlabels_2,testdata_2, testlabels_2, Deg_val2, labels2,
          traindata_3,trainlabels_3,testdata_3, testlabels_3, Deg_val3, labels3,
          traindata_4,trainlabels_4,testdata_4, testlabels_4, Deg_val4, labels4,
          traindata_5,trainlabels_5,testdata_5, testlabels_5, Deg_val5, labels5):
    prediction_1 = PolySVM(traindata_1,trainlabels_1,testdata_1, testlabels_1, Deg_val1) # First partition predictions
    prediction_2 = PolySVM(traindata_2,trainlabels_2,testdata_2, testlabels_2, Deg_val2) # Second partition predictions
    prediction_3 = PolySVM(traindata_3,trainlabels_3,testdata_3, testlabels_3, Deg_val3) # Third partition predictions
    prediction_4 = PolySVM(traindata_4,trainlabels_4,testdata_4, testlabels_4, Deg_val4) # Fourth partition predictions
    prediction_5 = PolySVM(traindata_5,trainlabels_5,testdata_5, testlabels_5, Deg_val5) # Fifth partition predictions

    errors = list()
    errors.append(calculateError(prediction_1,testlabels_1))     # Errors for first partition
    errors.append(calculateError(prediction_2,testlabels_2))     # Errors for second partition
    errors.append(calculateError(prediction_3,testlabels_3))     # Errors for third partition
    errors.append(calculateError(prediction_4,testlabels_4))     # Errors for fourth partition
    errors.append(calculateError(prediction_5,testlabels_5))     # Errors for fifth partition

    min_error = np.min(errors)
    min_pred = errors.index(min(errors))
    if(min_pred==0):
        best_Deg = Deg_val1
    if(min_pred==1):
        best_Deg = Deg_val2
    if(min_pred==2):
        best_Deg = Deg_val3
    if(min_pred==3):
        best_Deg = Deg_val4
    if(min_pred==4):
        best_Deg = Deg_val5
    return min_error, best_Deg
      
    
Deg_val1 = 0.5
Deg_val2 = 1
Deg_val3 = 2
Deg_val4 = 3
Deg_val5 = 4

min_error, best_Deg = bestDeg(traindata_1,trainlabels_1,testdata_1, testlabels_1, Deg_val1, testlabels_1,
          traindata_2,trainlabels_2,testdata_2, testlabels_2, Deg_val2, testlabels_2,
          traindata_3,trainlabels_3,testdata_3, testlabels_3, Deg_val3, testlabels_3,
          traindata_4,trainlabels_4,testdata_4, testlabels_4, Deg_val4, testlabels_4,
          traindata_5,trainlabels_5,testdata_5, testlabels_5, Deg_val5, testlabels_5)

print("Best Degree from K-Fold = ", best_Deg)

# Run Poly classifier   
PolySVM(X_tilde_train, Y_train, X_tilde_test, Y_test, best_Deg)    

# Save Poly classifier predictions
Y_pred_Poly = np.array([PolySVM(X_tilde_train, Y_train, X_tilde_test, Y_test, best_Deg)])

# Accuracy of the predicted labels
acc_label2 = np.equal(Y_pred_Poly,Y_test)

# Plot training data
plt.figure(figsize=(10,10))
setosa = plt.scatter(X_tilde_train.T[np.where(Y_train==0),0], X_tilde_train.T[np.where(Y_train==0),1],c='r',marker='o')
versicolor = plt.scatter(X_tilde_train.T[np.where(Y_train==1),0],X_tilde_train.T[np.where(Y_train==1),1],c='g',marker='o')
virginica = plt.scatter(X_tilde_train.T[np.where(Y_train==2),0],X_tilde_train.T[np.where(Y_train==2),1],c='b',marker='o')
plt.xlabel("Principal Component 1", fontsize = 15)
plt.ylabel("Principal Component 2", fontsize = 15)
plt.title("Two-dimensional PCA", fontsize = 15)

# Plot for test data
correct = plt.scatter(X_tilde_test.T[np.where(acc_label2==True),0],X_tilde_test.T[np.where(acc_label2==True),1],c='y',marker='s')
incorrect = plt.scatter(X_tilde_test.T[np.where(acc_label2==False),0],X_tilde_test.T[np.where(acc_label2==False),1],c='k',marker='^')
plt.legend((setosa, versicolor, virginica, correct, incorrect), 
               ("Setosa", "Versicolor", "Virginica", "Correctly Classified", "Incorrectly Classified"))
plt.show()
```

### Gaussian SVM on Iris Dataset

```python
def GaussianSVM(X_tilde_train, Y_train, X_tilde_test, Y_test, C_val):
    """
    Gaussian SVM classifier that returns either label "0", "1", or "2"
    Inputs:
        X_train: Training data matrix
        Y_train: Labels of training data matrix
        X_test: Test data matrix
        Y_test: Labels of test data matrix
    Outputs: 
        Y_pred: Predicted label of test data matrix
    """
    svclassifier = SVC(C = C_val, kernel = 'rbf')             # Gaussian SVM
    svclassifier.fit(X_tilde_train.T, Y_train)                # Fit the data
    Y_pred_Gaussian = svclassifier.predict(X_tilde_test.T)    # Make predictions on test data
    return Y_pred_Gaussian


def bestCG(traindata_1,trainlabels_1,testdata_1, testlabels_1, c1, labels1,
           traindata_2,trainlabels_2,testdata_2, testlabels_2, c2, labels2,
           traindata_3,trainlabels_3,testdata_3, testlabels_3, c3, labels3,
           traindata_4,trainlabels_4,testdata_4, testlabels_4, c4, labels4,
           traindata_5,trainlabels_5,testdata_5, testlabels_5, c5, labels5):
    prediction_1 = GaussianSVM(traindata_1,trainlabels_1,testdata_1, testlabels_1, c1) # First partition predictions
    prediction_2 = GaussianSVM(traindata_2,trainlabels_2,testdata_2, testlabels_2, c2) # Second partition predictions
    prediction_3 = GaussianSVM(traindata_3,trainlabels_3,testdata_3, testlabels_3, c3) # Third partition predictions
    prediction_4 = GaussianSVM(traindata_4,trainlabels_4,testdata_4, testlabels_4, c4) # Fourth partition predictions
    prediction_5 = GaussianSVM(traindata_5,trainlabels_5,testdata_5, testlabels_5, c5) # Fifth partition predictions

    errors = list()
    errors.append(calculateError(prediction_1,testlabels_1))     # Errors for first partition
    errors.append(calculateError(prediction_2,testlabels_2))     # Errors for second partition
    errors.append(calculateError(prediction_3,testlabels_3))     # Errors for third partition
    errors.append(calculateError(prediction_4,testlabels_4))     # Errors for fourth partition
    errors.append(calculateError(prediction_5,testlabels_5))     # Errors for fifth partition

    min_error = np.min(errors)
    min_pred = errors.index(min(errors))
    if(min_pred==0):
        best_C = c1
    if(min_pred==1):
        best_C = c2
    if(min_pred==2):
        best_C = c3
    if(min_pred==3):
        best_C = c4
    if(min_pred==4):
        best_C = c5
    return min_error, best_C
      
    
c1 = 1
c2 = 5
c3 = 10
c4 = 20
c5 = 100

min_error, best_C = bestCG(traindata_1,trainlabels_1,testdata_1, testlabels_1, c1, testlabels_1,
                           traindata_2,trainlabels_2,testdata_2, testlabels_2, c2, testlabels_2,
                           traindata_3,trainlabels_3,testdata_3, testlabels_3, c3, testlabels_3,
                           traindata_4,trainlabels_4,testdata_4, testlabels_4, c4, testlabels_4,
                           traindata_5,trainlabels_5,testdata_5, testlabels_5, c5, testlabels_5)
 
print("Best C = ", best_C)

# Save Gaussian classifier predictions
Y_pred_Gaussian = np.array([GaussianSVM(X_tilde_train, Y_train, X_tilde_test, Y_test, best_C)])

# Accuracy of the predicted labels
acc_label3 = np.equal(Y_pred_Gaussian,Y_test)

# Plot training data
plt.figure(figsize=(10,10))
setosa = plt.scatter(X_tilde_train.T[np.where(Y_train==0),0], X_tilde_train.T[np.where(Y_train==0),1],c='r',marker='o')
versicolor = plt.scatter(X_tilde_train.T[np.where(Y_train==1),0],X_tilde_train.T[np.where(Y_train==1),1],c='g',marker='o')
virginica = plt.scatter(X_tilde_train.T[np.where(Y_train==2),0],X_tilde_train.T[np.where(Y_train==2),1],c='b',marker='o')
plt.xlabel("Principal Component 1", fontsize = 15)
plt.ylabel("Principal Component 2", fontsize = 15)
plt.title("Two-dimensional PCA", fontsize = 15)

# Plot for test data
correct = plt.scatter(X_tilde_test.T[np.where(acc_label3==True),0],X_tilde_test.T[np.where(acc_label3==True),1],c='y',marker='s')
incorrect = plt.scatter(X_tilde_test.T[np.where(acc_label3==False),0],X_tilde_test.T[np.where(acc_label3==False),1],c='k',marker='^')
plt.legend((setosa, versicolor, virginica, correct, incorrect), 
               ("Setosa", "Versicolor", "Virginica", "Correctly Classified", "Incorrectly Classified"))
plt.show()
```

### Discussion of various types of SVM

The Setosa samples are linearly separable from from the Versicolor and the Virginica samples. Because of this linear separation, the Setosa are perfectly classified with linear, Gaussian, and polynomial SVM. Linear SVM is the best choice for classifying Setosa from Versicolor and Virginica because linear SVM is the faster and easiest classification, computationally wise.

The Versicolor and the Virginica samples are not linearly separable so they cannot be perfectly classified using linear SVM. Linear SVM is favored above Gaussian and polynomial SVM when the number of features is large. The reason for this is because if the number of features is large, Gaussian and polynomial SVM aren't significantly more accurate than linear SVM so the computational cost doesn't justify choosing Gaussian or polynomial over linear SVM. 

But in this example, there are only 2 principle component directions or features. So it is computationally worth using Gaussian and polynomial SVM instead of linear SVM. As seen above, Gaussian and polynomial SVM have slightly less classification errors than linear SVM.

# SVM on NBA Player Stats

### Importing NBA Player Stats Dataset

```python
import numpy as np
from sklearn import datasets 
from sklearn import svm
from sklearn.svm import SVC 
from sklearn.model_selection import train_test_split  
from sklearn.metrics import classification_report, confusion_matrix  
from matplotlib import pyplot as plt
import csv
import math

with open('players_stats_other.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter = ',')  
    count = -1
    number = []; playerNames = []; positions = []; points = []; fgPCT = []; threePCT = []; fthPCT = [];
    totalReb = [];  assist = []; steals = []; blocks = []; turnover = []; offReb = []; asstTO = []; labels = []
    
    for row in readCSV:
        count += 1
        number.append(count)
        playerNames.append(row[0])      # Use player names from 1st column in CSV dataset
        positions.append(row[30])       # Use positions from 29th column in CSV dataset
        points.append(row[3])           # Use points from 4th column in CSV dataset
        fgPCT.append(row[6])            # Use field goal percentage from 5th column in CSV dataset
        threePCT.append(row[9])         # Use 3 point percentage from 8th column in CSV dataset
        fthPCT.append(row[12])          # Use free throw percentage from 11th column in CSV dataset
        offReb.append(row[13])          # Use offensive rebounds from 12th column in CSV dataset
        totalReb.append(row[15])        # Use total rebounds from 14th column in CSV dataset
        assist.append(row[16])          # Use assists from 15th column in CSV dataset
        steals.append(row[17])          # Use steals from 16th column in CSV dataset
        blocks.append(row[18])          # Use blocks from 17th column in CSV dataset
        turnover.append(row[19])        # Use turnovers from 18th column in CSV dataset
        asstTO.append(row[22])          # Use assist-turnover ratio from 21st column in CSV dataset
        
        if(row[30] == 'PG'):
            labels.append(1)
        if(row[30] == 'SG'):
            labels.append(2)
        if(row[30] == 'SF'):
            labels.append(3)
        if(row[30] == 'PF'):
            labels.append(4)
        if(row[30] == 'C'):
            labels.append(5)    

                
numA = np.column_stack(number)
namesA = np.column_stack(playerNames)
posA = np.column_stack(positions)
pointsA = np.column_stack(points)
fgA = np.column_stack(fgPCT)
threeA = np.column_stack(threePCT)
freeA = np.column_stack(fthPCT)
rebA = np.column_stack(totalReb)
assistA = np.column_stack(assist)
stealA = np.column_stack(steals)
blockA = np.column_stack(blocks)
turnA = np.column_stack(turnover)
offRA = np.column_stack(offReb)
atoA = np.column_stack(asstTO)

statsNames = np.array(list(zip(numA, namesA, posA, pointsA, fgA, threeA, freeA, rebA, assistA, stealA, blockA, turnA)))
statsN = statsNames[0, :,:]

statsNoNames = np.array(list(zip(fgA, atoA, threeA, freeA, rebA, pointsA, assistA, stealA, blockA, turnA, offRA)))
stats = statsNoNames[0, :,:]

data = stats.T[1:,:]
names = playerNames[1:]
pos = positions[1:]
dataSet = data.astype(np.float)
```

### PCA and Plot of NBA Dataset

```python
length_train = (int)(490*0.9)            # 90% of data used to train
length_test = (int)(490*0.1)             # 10% of data used to test

X_train = dataSet[0:length_train,:]      # Splitting dataset into train and test
Y_train = labels[0:length_train]
X_test = dataSet[length_train:,:]
Y_test = labels[length_train:]


# Computation of the mean of training data
X_train_T = X_train.T
mean_vec = X_train_T.mean(axis = 1)
mean_mat = np.tile(mean_vec.reshape(11,1), (1,length_train))

# Centered training data matrix
train_centered_X = X_train_T - mean_mat

# Normalize training data matrix
standard_deviation = np.array([[np.std(i)] for i in train_centered_X])
train_normal_X = train_centered_X/standard_deviation

# Compute two-dimensional PCA features of the centered and normalized dataset
U, s, Vh = np.linalg.svd(train_normal_X)    # Singular value decomposition 

# Matrix of the top two principal components of the centered, normalized training data matrix
U_2 = U[:,0:2]

# PCA-based features of the centered, normalized data using the top-two principal components
X_tilde_train = U_2.T@train_normal_X

# Computation of the mean of test data
X_test_T = X_test.T
mean_mat = np.tile(mean_vec.reshape(11,1), (1,length_test))

# Centered test data matrix
test_centered_X = X_test_T - mean_mat

# Normalize test data matrix
standard_deviation = np.array([[np.std(i)] for i in test_centered_X])
test_normal_X = test_centered_X/standard_deviation

# PCA-based features of the centered, normalized data using the top-two principal components
X_tilde_test = U_2.T@test_normal_X


# Scatter plot of training data using the two-dimensional PCA features
plt.figure(figsize=(15,15))

oneX = list();   oneY = list();
twoX = list();   twoY = list();
threeX = list(); threeY = list();
fourX =  list(); fourY = list();
fiveX = list();  fiveY = list();

for i in range(len(X_train)):
    if(labels[i] == 1):
        oneX.append(X_tilde_train.T[i][0])
        oneY.append(X_tilde_train.T[i][1])
    if(labels[i] == 2):
        twoX.append(X_tilde_train.T[i][0])
        twoY.append(X_tilde_train.T[i][1])
    if(labels[i] == 3):
        threeX.append(X_tilde_train.T[i][0])
        threeY.append(X_tilde_train.T[i][1])
    if(labels[i] == 4):
        fourX.append(X_tilde_train.T[i][0])
        fourY.append(X_tilde_train.T[i][1])
    if(labels[i] == 5):
        fiveX.append(X_tilde_train.T[i][0])
        fiveY.append(X_tilde_train.T[i][1])
            

PG = plt.scatter(oneX, oneY, c='m', marker='o', label = 'PG')
SG = plt.scatter(twoX, twoY, c='k',marker='s', label = 'SG')
SF = plt.scatter(threeX, threeY, c='b',marker='^', label = 'SF')
PF = plt.scatter(fourX, fourY, c='y',marker='D', label = 'PF')
C  = plt.scatter(fiveX, fiveY, c='c',marker='s', label = 'C')
plt.legend(handles = [PG, SG, SF, PF, C])
plt.xlabel("Principal Component 1", fontsize = 15)
plt.ylabel("Principal Component 2", fontsize = 15)
plt.title("Two-dimensional PCA of NBA Dataset", fontsize = 15)
plt.show()
```

### K-Fold on NBA Data

```python
k = round(441/5)                      # Round to find length/5 for K=5
f = len(X_tilde_train.T)              # Length of samples

traindata_1 = X_tilde_train[:,k:f]    # First partition
trainlabels_1 = Y_train[k:f]
testdata_1 = X_tilde_train[:,0:k]
testlabels_1 = Y_train[0:k]

traindata_2 = np.column_stack((X_tilde_train[:,0:k],X_tilde_train[:,(2*k):]))      # Second partition
trainlabels_2 = np.append(Y_train[:k],Y_train[(2*k):])
testdata_2 = X_tilde_train[:,k:(2*k)]
testlabels_2 = Y_train[k:(2*k)]

traindata_3 = np.column_stack((X_tilde_train[:,0:(2*k)],X_tilde_train[:,(3*k):]))  # Third partition
trainlabels_3 = np.append(Y_train[:(2*k)],Y_train[(3*k):])
testdata_3 = X_tilde_train[:,(2*k):(3*k)]
testlabels_3 = Y_train[(2*k):(3*k)]

trainlabels_4 = np.append(Y_train[:(3*k)],Y_train[(4*k):])     # Fourth partition
traindata_4 = np.column_stack((X_tilde_train[:,0:(3*k)],X_tilde_train[:,(4*k):]))
testdata_4 = X_tilde_train[:,(3*k):(4*k)]
testlabels_4 = Y_train[(3*k):(4*k)]

trainlabels_5 = Y_train[:(4*k)]                  # Fifth partition
traindata_5 = X_tilde_train[:,0:(4*k)]
testdata_5 = X_tilde_train[:,(4*k):]
testlabels_5 = Y_train[(4*k):]

def calculateError(pred,labels):
    count = 0;
    for i in range(len(pred)):
        if (pred[i] != labels[i]):
            count = count +1
    percentage = count/(len(pred))
    return percentage
```

### Linear SVM of NBA Dataset

```python
def LinearSVM(X_tilde_train, Y_train, X_tilde_test, Y_test, C_val):
    """
    Linear SVM classifier that returns either label "0", "1", or "2"
    Inputs:
        X_train: Training data matrix
        Y_train: Labels of training data matrix
        X_test: Test data matrix
        Y_test: Labels of test data matrix
    Outputs: 
        Y_pred: Predicted label of test data matrix
    """
    svclassifier = SVC(C = C_val, kernel = 'linear')        # Linear SVM
    svclassifier.fit(X_tilde_train.T, Y_train)              # Fit the data
    Y_pred_Linear = svclassifier.predict(X_tilde_test.T)    # Make predictions on test data
    return Y_pred_Linear

def bestC(traindata_1,trainlabels_1,testdata_1, testlabels_1, c1, labels1,
          traindata_2,trainlabels_2,testdata_2, testlabels_2, c2, labels2,
          traindata_3,trainlabels_3,testdata_3, testlabels_3, c3, labels3,
          traindata_4,trainlabels_4,testdata_4, testlabels_4, c4, labels4,
          traindata_5,trainlabels_5,testdata_5, testlabels_5, c5, labels5):
    prediction_1 = LinearSVM(traindata_1,trainlabels_1,testdata_1, testlabels_1, c1) # First partition predictions
    prediction_2 = LinearSVM(traindata_2,trainlabels_2,testdata_2, testlabels_2, c2) # Second partition predictions
    prediction_3 = LinearSVM(traindata_3,trainlabels_3,testdata_3, testlabels_3, c3) # Third partition predictions
    prediction_4 = LinearSVM(traindata_4,trainlabels_4,testdata_4, testlabels_4, c4) # Fourth partition predictions
    prediction_5 = LinearSVM(traindata_5,trainlabels_5,testdata_5, testlabels_5, c5) # Fifth partition predictions

    errors = list()
    errors.append(calculateError(prediction_1,testlabels_1))     # Errors for first partition
    errors.append(calculateError(prediction_2,testlabels_2))     # Errors for second partition
    errors.append(calculateError(prediction_3,testlabels_3))     # Errors for third partition
    errors.append(calculateError(prediction_4,testlabels_4))     # Errors for fourth partition
    errors.append(calculateError(prediction_5,testlabels_5))     # Errors for fifth partition

    min_error = np.min(errors)
    min_pred = errors.index(min(errors))
    if(min_pred==0):
        best_C = c1
    if(min_pred==1):
        best_C = c2
    if(min_pred==2):
        best_C = c3
    if(min_pred==3):
        best_C = c4
    if(min_pred==4):
        best_C = c5
    return min_error, best_C
      
    
c1 = 1
c2 = 5
c3 = 10
c4 = 20
c5 = 100

min_error, best_C = bestC(traindata_1,trainlabels_1,testdata_1, testlabels_1, c1, testlabels_1,
                          traindata_2,trainlabels_2,testdata_2, testlabels_2, c2, testlabels_2,
                          traindata_3,trainlabels_3,testdata_3, testlabels_3, c3, testlabels_3,
                          traindata_4,trainlabels_4,testdata_4, testlabels_4, c4, testlabels_4,
                          traindata_5,trainlabels_5,testdata_5, testlabels_5, c5, testlabels_5)
 
print("Best C form K-Fold = ", best_C)

# Save Linear classifier predictions
Y_pred_Linear = np.array([LinearSVM(X_tilde_train, Y_train, X_tilde_test, Y_test, best_C)])

# Accuracy of the predicted labels
acc_label1 = np.equal(Y_pred_Linear,Y_test)

# Plot training data
# Plot training data
plt.figure(figsize=(15,15))
oneX = list();   oneY = list();
twoX = list();   twoY = list();
threeX = list(); threeY = list();
fourX =  list(); fourY = list();
fiveX = list();  fiveY = list();

for i in range(len(X_train)):
    if(labels[i] == 1):
        oneX.append(X_tilde_train.T[i][0])
        oneY.append(X_tilde_train.T[i][1])
    if(labels[i] == 2):
        twoX.append(X_tilde_train.T[i][0])
        twoY.append(X_tilde_train.T[i][1])
    if(labels[i] == 3):
        threeX.append(X_tilde_train.T[i][0])
        threeY.append(X_tilde_train.T[i][1])
    if(labels[i] == 4):
        fourX.append(X_tilde_train.T[i][0])
        fourY.append(X_tilde_train.T[i][1])
    if(labels[i] == 5):
        fiveX.append(X_tilde_train.T[i][0])
        fiveY.append(X_tilde_train.T[i][1])


PG = plt.scatter(oneX, oneY, c='m', marker='o', s = 50, alpha = 0.5)
SG = plt.scatter(twoX, twoY, c='k',marker='s', s = 50, alpha = 0.5)
SF = plt.scatter(threeX, threeY, c='b',marker='^', s = 50, alpha = 0.5)
PF = plt.scatter(fourX, fourY, c='y',marker='D', s = 50, alpha = 0.5)
C  = plt.scatter(fiveX, fiveY, c='c',marker='s', s = 50, alpha = 0.5)
correct = plt.scatter(X_tilde_test.T[np.where(acc_label1==True),0],X_tilde_test.T[np.where(acc_label1==True),1],c='g',marker='o', linewidths=3.5, s=100)
incorrect = plt.scatter(X_tilde_test.T[np.where(acc_label1==False),0],X_tilde_test.T[np.where(acc_label1==False),1],c='r',marker='o', linewidths=3.5, s = 100)
plt.legend((PG, SG, SF, PF, C, correct, incorrect),("PG", "SG", "SF", "PF", "C", "Classified Correctly", "Incorrectly Classified"))
plt.xlabel("Principal Component 1", fontsize = 15)
plt.ylabel("Principal Component 2", fontsize = 15)
plt.title("Two-dimensional PCA", fontsize = 15)
plt.show()
```

### Discussion on NBA SVM

As stated above, linear SVM is a wise choice on datasets with a large number of features. I performed 2D PCA on the NBA dataset. Because I limited the number of features to only 2, it is computationally worth using polynomial or Gaussian SVM on this dataset. 

For linear, polynomial, and Gaussian SVM, each had about 65 to 70% correct results. This low accuracy rate could be a result of several things. First, several NBA players can play multiple positions, but all players in the dataset were only given 1 position. So a tall power forward could be label as just a power forward, but in some games he could actually be playing the center position. 

Another reason for misclassification are outliers. Superstars are outliers because they are usually very well rounded. A superstar point guard will probably have more blocks and rebounds than average point guards which skews the superstars data. Another type of outlier is a player who doesn't play many minutes in games. These players might only play for a couple minutes and therefore can have skewed averages of statistics. 

In the graph of the normalized data, the outliers to the left of the plot represent players with statistics that vary a lot. These players are most likely players who don't play a lot of minutes so their averages have high variance. The data points to the right of on the normalized graph most likely represent superstars because their statistical variance is very low, meaning they are well rounded players.

```python
length_train = (int)(210*0.8)       # 80% of data for training
length_test = (int)(210*0.2)        # 30% of data for testing

X_train = dataSet[:,0:length_train]
X_test = dataSet[:,length_train:]

# Computation of the mean of training data
mean_vec = X_train.mean(axis = 1)
mean_mat = np.tile(mean_vec.reshape(7,1), (1,length_train))

# Centered training data matrix
train_centered_X = X_train - mean_mat

# Normalize training data matrix
standard_deviation = np.array([[np.std(i)] for i in train_centered_X])
train_normal_X = train_centered_X/standard_deviation

# Compute two-dimensional PCA features of the centered and normalized dataset
U, s, Vh = np.linalg.svd(train_normal_X)    # Singular value decomposition 

# Matrix of the top two principal components of the centered, normalized training data matrix
U_2 = U[:,0:2]

# PCA-based features of the centered, normalized data using the top-two principal components
X_tilde_train = U_2.T@train_normal_X

# Computation of the mean of test data
mean_vec = X_test.mean(axis=1)
mean_mat = np.tile(mean_vec.reshape(7,1), (1,length_test))

# Centered test data matrix
test_centered_X = X_test - mean_mat

# Normalize test data matrix
standard_deviation = np.array([[np.std(i)] for i in test_centered_X])
test_normal_X = test_centered_X/standard_deviation

# PCA-based features of the centered, normalized data using the top-two principal components
X_tilde_test = U_2.T@test_normal_X

def PlotTrainingDataSeed(X_tilde_train_T):
    plt.figure(figsize=(10,10))
    plt.scatter(X_tilde_train[0], X_tilde_train[1],c='r',marker='o')
    plt.xlabel("Principal Component 1", fontsize = 15)
    plt.ylabel("Principal Component 2", fontsize = 15)
    plt.title("Two-dimensional PCA", fontsize = 15)

PlotTrainingDataSeed(X_tilde_train)
```
