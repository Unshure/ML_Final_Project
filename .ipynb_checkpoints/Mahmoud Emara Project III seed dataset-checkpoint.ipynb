### Clustering Wheat Data

In this dataset, 210 randomly selected elements were analyzed based on three different types of wheat: Kama, Rose and Canadian. For each type of wheat, 70 elements were selected. This form of data is able to be used for clustering or classification because the dataset included a "target" vector for validation. In this exercise, we will be performing Agglomerative Hierarchical Clustering by ignoring the target vector and only using it at the end to measure the accuracy of clusters formed and for graphical presentation.

```python
import csv
import numpy as np
import math
from matplotlib import pyplot as plt
import scipy
import sklearn
from sklearn.model_selection import train_test_split

with open(r'Seed_Data.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    area = []; perimeter = []
    compactness = []; lengKern = []
    widKern = []; asymcoeff = []
    lenKernGroove = []; target = []
    
    for row in readCSV:
        area.append(row[0])
        perimeter.append(row[1])
        compactness.append(row[2])
        lengKern.append(row[3])
        widKern.append(row[4])
        asymcoeff.append(row[5])
        lenKernGroove.append(row[6])
        target.append(row[7])

areaA = np.column_stack(area)
permA = np.column_stack(perimeter)
compA = np.column_stack(compactness)
lkerA = np.column_stack(lengKern)
wkerA = np.column_stack(widKern)
ascoA = np.column_stack(asymcoeff)
grovA = np.column_stack(lenKernGroove)
targA = np.column_stack(target)

allData= np.array(list(zip(areaA, permA, compA, lkerA, wkerA, ascoA, grovA, targA)))
primData = allData[0,:,1:].T
dataSet = allData[0,0:7,1:].T
targetVal =allData[0,7:8,1:].T

primData = primData.astype(np.float)
dataSet = dataSet.astype(np.float)
targetVal = targetVal.astype(np.float)

# Will not be using any test data for clustering problem.
splitR = 0
limit = (int)((1-splitR)*len(dataSet))
trainSet, testSet, ytrain, ytest = train_test_split(dataSet, targetVal, test_size = splitR)

print("Data dimensions:",dataSet.shape,"\nTrain Set dimensions:",trainSet.shape)
```

```python
# PCA on training dataset for graphical presentation at end of code.
trainingSet = trainSet.T
meanVector = trainingSet.mean(axis=1)
meanMat = np.tile(meanVector.reshape(7,1),(1,limit))
centData = trainingSet - meanMat
print("Centered Data dimensions: ",centData.shape)

U, s, Vt = np.linalg.svd(centData)
kComp = 2
princ = U[:,0:kComp]
Xtilde = np.matmul(princ.T, centData)

# Normalization of Data
normData = np.zeros((7,limit))

for i in range(len(trainingSet)):
    std = np.std(centData[i])
    for j in range(len(trainingSet[0])):
        normData[i][j] = centData[i][j] / std

print("Normalized data dimensions: ", normData.shape)

U, s, Vt = np.linalg.svd(normData)
newPrinc = U[:,0:kComp]
Xhat = np.matmul(newPrinc.T, normData)
print("Dimensions of projected data:",Xhat.shape)

plt.figure(figsize=(8,6), dpi= 80, facecolor='w', edgecolor='k')
onePlot = plt.scatter(Xhat[0], Xhat[1], color='b', marker='o', label='Data')
plt.title('Two-Dimensional Features of Normalized Wheat Data')
plt.xlabel('Principal compenent 1')
plt.ylabel('Principal compenent 2')
plt.show()
```

```python
import random 

def coordinatesN(data, label, l):
    xcoord = list(); ycoord = list()
    for i in range(len(data)):
        if(label[i] == l):
            xcoord.append(Xhat.T[i][0])
            ycoord.append(Xhat.T[i][1])
    return xcoord, ycoord

zerX1, zerY1 = coordinatesN(trainSet, ytrain, 0)
oneX1, oneY1 = coordinatesN(trainSet, ytrain, 1)
twoX1, twoY1 = coordinatesN(trainSet, ytrain, 2)

plt.figure(figsize=(8,6), dpi= 80, facecolor='w', edgecolor='k')
zerPlot1 = plt.scatter(zerX1, zerY1, color='r', marker='o', label='Kama')
onePlot1 = plt.scatter(oneX1, oneY1, color='g', marker='o', label='Rose')
twoPlot1 = plt.scatter(twoX1, twoY1, color='b', marker='o', label='Canadian')
plt.title('2D Features of Normalized Data')
plt.xlabel('Principal compenent 1')
plt.ylabel('Principal compenent 2')
plt.legend(handles = [zerPlot1, onePlot1, twoPlot1], loc='upper left')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import fcluster
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist

from pylab import rcParams
import seaborn as sb

from sklearn.cluster import AgglomerativeClustering
import sklearn.metrics as sm


# Generate Dendrogram
Z = linkage(trainSet, 'ward')
plt.figure(figsize=(10,8))
dendrogram(Z, truncate_mode='lastp', p=7, leaf_rotation=0., leaf_font_size=12., show_contracted=True)
plt.title('Truncated Hierarchal Clustering Dendrogram')
plt.xlabel('Cluster Size')
plt.ylabel('Distance')
plt.axhline(y=24) # Helps determine number of clusters formed at minimum distance.
plt.grid()
plt.show()
```

The variable Z represents the instantation of the linkage method. The linkage method determines which distance to use when monitoring the data. If we use "ward", it would minimize the variance of the clusters (or the minimal increase of sum of squares). "Complete" would  would use the maximum distance between all data and "average would take the average distances. Other methods include Median, Centroid and Single. When generating the dendrogram, we want to make sure it is easy to read, so some parameters were filled in to make it clear. We use "lastp" truncate mode to correspond to the only non-leaf nodes in the linkage.

As we plot the dendrogram we see that 32 data points and 15 points were grouped as a cluster, and a cluster of 12 and 60 points were formed as well. The sum of these values should equal the dataset itself as it represents the data as a whole. The y-axis represents the distance at which different clusters were formed. The horizontal lines give an easy view of where the clusters form, at distance = 24, we see 3 clusters form.

Moving on, we try to instantiate all possible combinations of agglomerative clustering and choose the one that gives the best accuracy.

```python
# Instantiate an agglomerative clustering object and check the true values of labels.

k = 3 # 3 clusters
Hclustering = AgglomerativeClustering(n_clusters=k, affinity='euclidean', linkage='average')
trial1 = Hclustering.fit(trainSet)
sm.accuracy_score(ytrain, trial1.labels_)

# Affinity parameter represents similarity of data through type of distance calculated.
```

```python
Hclustering = AgglomerativeClustering(n_clusters=k, affinity='euclidean', linkage='complete')
trial2 = Hclustering.fit(trainSet)
sm.accuracy_score(ytrain, trial2.labels_)
```

```python
Hclustering = AgglomerativeClustering(n_clusters=k, affinity='euclidean', linkage='ward')
trial3 = Hclustering.fit(trainSet)
sm.accuracy_score(ytrain, trial3.labels_)
```

```python
Hclustering = AgglomerativeClustering(n_clusters=k, affinity='manhattan', linkage='average')
trial4 = Hclustering.fit(trainSet)
sm.accuracy_score(ytrain, trial4.labels_)
```

```python
Hclustering = AgglomerativeClustering(n_clusters=k, affinity='manhattan', linkage='complete')
trial5 = Hclustering.fit(trainSet)
sm.accuracy_score(ytrain, trial5.labels_)
```

```python
Hclustering = AgglomerativeClustering(n_clusters=k, affinity='cosine', linkage='average')
trial6 = Hclustering.fit(trainSet)
sm.accuracy_score(ytrain, trial6.labels_)
```

```python
Hclustering = AgglomerativeClustering(n_clusters=k, affinity='cosine', linkage='complete')
trial7 = Hclustering.fit(trainSet)
sm.accuracy_score(ytrain, trial7.labels_)
```

We see that the best score was performed when the distance metric used was Manhattan distance, and a linkage parameter of Average.

```python
# print(Hclustering.labels_)
# print(ytrain)

predc0x = []; predc1x = []; predc2x = []
predc0y = []; predc1y = []; predc2y = []

for i in range(len(trainSet)):
    if(trial4.labels_[i] == 0):
        predc0x.append(Xhat.T[i][0])
        predc0y.append(Xhat.T[i][1])
    if(trial4.labels_[i] == 1):
        predc1x.append(Xhat.T[i][0])
        predc1y.append(Xhat.T[i][1])
    if(trial4.labels_[i] == 2):
        predc2x.append(Xhat.T[i][0])
        predc2y.append(Xhat.T[i][1])

plt.figure(figsize=(10,8), dpi= 80, facecolor='w', edgecolor='k')
pzPlot = plt.scatter(predc0x, predc0y, color='r', marker='x', label='Kama')
poPlot = plt.scatter(predc1x, predc1y, color='g', marker='x', label='Rose')
ptPlot = plt.scatter(predc2x, predc2y, color='b', marker='x', label='Canadian')
plt.title('Clustered Wheat Data')
plt.xlabel('Principal compenent 1')
plt.ylabel('Principal compenent 2')
plt.legend(handles = [pzPlot, poPlot, ptPlot], loc='upper left')
plt.show()
    
```
